```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn . metrics import accuracy_score
```

```python
iris = load_iris()
dir(iris)
```

```python
dir(iris)
```

```python
df=sns.load_dataset("iris")
df.head()
```

```python
X = df.drop(["species"],axis="columns")
```

```python
y = iris.target
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
treemodel = DecisionTreeClassifier(random_state=42)
treemodel.fit(X_train,y_train)
```

```python
y_pred = treemodel.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
depth = treemodel.get_depth()
leaves = treemodel.get_n_leaves()
```

```python
print(f"DT Accuracy : {accuracy*100:.2f}")
print(f"Tree Depth: {depth}")
print(f"Tree Leaves: {leaves}")
```

```python
class_names = ["setosa","versicolor","verginica"]
plt.figure(figsize=(16,8))
plot_tree(treemodel,filled=True,feature_names=X.columns,class_names=class_names,fontsize=8)
plt.show()
```
