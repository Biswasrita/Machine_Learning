```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
import seaborn as sns
from sklearn.linear_model import LogisticRegression
```

```python
iris = load_iris()
dir(iris)
```

```python
df = sns.load_dataset("iris")
df
```

```python
X = df.drop(["species"],axis = "columns")
X
```

```python
y = iris.target
```

```python
model = LogisticRegression(solver='lbfgs',max_iter=200,multi_class='multinomial')
model.fit(X,y)
y_pred_multi = model.predict(X)
accuracy = accuracy_score (y,y_pred_multi)
cr = classification_report (y,y_pred_multi)
cm = confusion_matrix (y,y_pred_multi)
```

```python
print("Multiclass Logistic Regression")
print(f"Confusion matrix:\n{cm}")
print(f"Classification Report:\n{cr}")
print(f"Accuracy:\n{accuracy*100:.2f}")
sns.heatmap(cm,annot=True)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
```

```python
binary_filter = y<2
X_binary = X[binary_filter]
y_binary = y[binary_filter]
model = LogisticRegression()
model.fit(X_binary,y_binary)
```

```python
y_pred_binary = model.predict(X_binary)
accuracy1 = accuracy_score (y_binary,y_pred_binary)
cr1 = classification_report (y_binary,y_pred_binary)
cm1 = confusion_matrix (y_binary,y_pred_binary)
```

```python
print("Binary Logistic Regression")
print(f"Confusion matrix:\n{cm1}")
print(f"Classification Report:\n{cr1}")
print(f"Accuracy:\n{accuracy1*100:.2f}")
sns.heatmap(cm1,annot=True)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
```
